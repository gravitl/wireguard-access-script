```python
from python_wireguard import Key
import os
import time
def generate_key_pair():
    private_key, public_key = Key.key_pair()
    return private_key, public_key
##################################################################################################################################################################
def create_config(peer_num, private_key, address):
    config = f"""
[Interface]
PrivateKey = {private_key}
Address = {address}
DNS = 10.101.0.6

[Peer]
PublicKey = kTA3Ci7XskXHUd2v6/HOoZgRq0GsRNabBkv1jSPZ1j4=
Endpoint = 146.190.74.23:51820
AllowedIPs = 10.191.143.0/24, 10.101.0.0/16
PersistentKeepalive = 25
"""
    peer_num = peer_num -1
    filename = os.path.join('/content/drive/MyDrive/wg_config files',  f'wg_peer_{peer_num}.conf')
    with open(filename, 'w') as file:
        file.write(config)
    print(f"Configuration file for peer {peer_num} created: {filename}")


################################################################################################################################################################
def create_gateway_config(peers_public_keys):

    config = f"""
[Interface]
PrivateKey = 0Jy8eUum5qYGrSIlVF/+RGj505EjiXiE1yQms7/Sl2U=
Address = 10.191.143.1/32
ListenPort = 51820

PostUp = sysctl -w net.ipv4.ip_forward=1
PostUp = nft add rule ip nat postrouting oifname "eth1" masquerade
PostDown = nft delete rule ip nat postrouting oifname "eth1" masquerade
"""
    for i, public_key in enumerate(peers_public_keys, start=2):
        config += f"""
[Peer]
PublicKey = {public_key}
AllowedIPs = 10.191.143.{i}/32
PersistentKeepalive = 25
"""
    filename = os.path.join('/content/drive/MyDrive/wg_config files', 'wg_gateway.conf')
    with open(filename, 'w') as file:
        file.write(config)
    print(f"Gateway configuration file created: {filename}")
####################################################################################################################################################################

def main():
    num_peers = int(input("Enter the number of peers: "))

    peers_public_keys = []
    for i in range(2, num_peers + 2):
        private_key, public_key = generate_key_pair()
        address = f"10.191.143.{i}/32"
        peers_public_keys.append(public_key)
        create_config(i, private_key, address)
        time.sleep(10)


    create_gateway_config(peers_public_keys)

    print("Done")

if __name__ == "__main__":
    main()
```
